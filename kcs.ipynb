```python
import pandas as pd    
path = "/cluster/work/cotterell/kdu/ml-knowledge-conflicts/datasets/substitution-sets/MRQANaturalQuestionsDevCorpusSub.jsonl"
kcs = pd.read_json(path_or_buf=path, lines=True)
```

```python
kcs = kcs.dropna(subset="query")
kcs = kcs.dropna(axis="columns")
```

```python
kcs["wh-question"] = kcs["query"].apply(lambda x: x.split()[0])
```

```python
from gensim import corpora, models
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import string

import nltk
nltk.download('punkt')
nltk.download('stopwords')

# Sample set of questions
questions = kcs[kcs["wh-question"] == "who"]["query"].unique()

# Tokenize and preprocess the text
stop_words = set(stopwords.words('english'))
punctuation = set(string.punctuation)
texts = [
    [word.lower() for word in word_tokenize(question) if word.lower() not in stop_words and word.lower() not in punctuation]
    for question in questions
]

# Create a dictionary representation of the documents
dictionary = corpora.Dictionary(texts)

# Convert the list of questions into a bag-of-words corpus
corpus = [dictionary.doc2bow(text) for text in texts]

# Train the LDA model
lda_model = models.LdaModel(corpus, num_topics=50, id2word=dictionary, passes=15)

# Print the topics and their most significant words
for idx, topic in lda_model.print_topics(-1):
    print(f"Topic {idx + 1}: {topic}")

# # To get topics for a new question
# new_question = "How does pollution affect marine life?"
# new_question_bow = dictionary.doc2bow(word_tokenize(new_question.lower()))
# topics = lda_model.get_document_topics(new_question_bow)
# print(f"Topics for the new question: {topics}")
```
