```python
import pandas as pd
import json
from measureLM import helpers
```

```python
def unpack_entities(entities, mention_type='auto_match'):
    
    ents = []
    for i in [0,1]:
        ent = entities[0][i][mention_type]
        if ent == "<Someone>":
            ent = "The speaker"
        if ent == "<Author>":
            ent = "The author"
        ents.append(ent)
            
    label = entities[0][2]['label']        
    return *ents, label


df = pd.read_json(helpers.ROOT_DIR / "data" / "SEESAW" / "train.json")

df[["ent1", "ent2", "label"]] = df.apply(lambda row: unpack_entities(row["entities"]), axis=1, result_type="expand")
df = df.drop(columns=['context text', 'document text', 'entities', 'unique_id'])
df
```
